# Training and Testing data

```python
#  Getting Features

x = data.drop(columns = 'diagnosis')

# Getting Predicting Value
y = data['diagnosis']
```

```python
#train_test_splitting of the dataset
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(len(x_train))
x_train
```

```python
print(len(x_test))
```

```python
print(len(y_train))
y_train
```

```python
print(len(y_test))
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# MODEL

## 1. SVM

```python
from sklearn.svm import SVC

#Initalize the classifier
classifier = SVC(kernel="rbf")

#Fitting the training data
classifier.fit(x_train, y_train)
```

```python
#Predicting on test
y_pred2 = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_squared_error
#Classification Report
print('\nClassification Report')
print(classification_report(y_test,y_pred2))

from sklearn.metrics import confusion_matrix, accuracy_score
#Confusion Matrix
print('\nConfusion matrix')
cm = confusion_matrix(y_test, y_pred2)
cm
```

```python
print("Accuracy ",accuracy_score(y_test, y_pred2)*100)
```

# 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

#Initalize the classifier
dtree = DecisionTreeClassifier(max_depth=6, random_state=123)

#Fitting the training data
dtree.fit(x_train,y_train)
```

```python
#Predicting on test
y_pred3=dtree.predict(x_test)
```

```python

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_squared_error

#Classification Report
print('\nClassification Report')
print(classification_report(y_test,y_pred3))

#Confusion Matrix
print('\nConfusion matrix')
print(confusion_matrix(y_test,y_pred3))
print("Training Score: ",dtree.score(x_train,y_train)*100)
```

```python
#Accuracy
print("Accuracy  ",accuracy_score(y_test,y_pred3)*100)
```

## 3. Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
#Initalize the classifier
rfc=RandomForestClassifier()

#Fitting the training data
rfc.fit(x_train,y_train)
```

```python
#Predicting on test
y_pred4=rfc.predict(x_test)
```

```python

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_squared_error
#Classification Report
print('\nClassification Report')
print(classification_report(y_test,y_pred4))

#Confusion Matrix
print('\nConfusion matrix')
print(confusion_matrix(y_test,y_pred4))
print("Training Score: ",rfc.score(x_train,y_train)*100)
```

```python
#Accuracy
print("Accuracy  ",accuracy_score(y_test,y_pred4)*100)
```

# 4. KNeighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
#Initalize the classifier
knn = KNeighborsClassifier(n_neighbors=15)
#Fitting the training data
knn.fit(x_train, y_train)
```

```python
#Predicting on test
y_pred5=knn.predict(x_test)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
print('\nConfusion matrix')
print(confusion_matrix(y_test, y_pred5))

#Classification Report
from sklearn.metrics import classification_report
print('\nClassification Report')
print(classification_report(y_test, y_pred5)) 
```

```python
#Accuracy
print('Accuracy = ', knn.score(x_test, y_test)*100)
```
